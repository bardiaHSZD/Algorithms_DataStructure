### **2. BFS Traversal in Python**

```python
class Graph:
  def __init__(self, gdict=None):
    if gdict is None:
      gdict={}
    self.gdict = gdict  
  
  def addEdge(self, vertex, edge):
    self.gdict[vertex].append(edge)    
  
  def bfs(self, vertex):
    visited = [vertex]
    queue = [vertex]
    while queue:
      deVertex = queue.pop(0)
      print(deVertex)
      for adjucentVertex in self.gdict[deVertex]:
        if adjucentVertex not in visited:
          visited.append(adjucentVertex)
          queue.append(adjucentVertex)


customDict = {"a": ["b", "c"],
              "b": ["a", "d", "e"],
              "c": ["a", "e"],
              "d": ["b", "e", "f"],
              "e": ["b", "c", "d"],
              "f": ["d", "e"]}

simple_unweighted_undirected_graph = Graph(customDict)
print(simple_unweighted_undirected_graph.gdict)
simple_unweighted_undirected_graph.addEdge(vertex="e", edge="f")
print(simple_unweighted_undirected_graph.gdict["e"])

print("Traversing Through the graph with BFS")
simple_unweighted_undirected_graph.bfs("a")
```

### **4. DFS Traversal in Python**

```python
class Graph:
  def __init__(self, gdict=None):
    if gdict is None:
      gdict={}
    self.gdict = gdict  
  
  def addEdge(self, vertex, edge):
    self.gdict[vertex].append(edge)    
  
  def bfs(self, vertex):
    visited = [vertex]
    queue = [vertex]
    while queue:
      deVertex = queue.pop(0) #first element
      print(deVertex)
      for adjucentVertex in self.gdict[deVertex]:
        if adjucentVertex not in visited:
          visited.append(adjucentVertex)
          queue.append(adjucentVertex)

  def dfs(self,vertex):
    visited = [vertex]
    stack = [vertex]
    while stack:
      popVertex = stack.pop() #last element       
      print(popVertex)
      for adjucentVertex in self.gdict[popVertex]:
        if adjucentVertex not in visited:
          visited.append(adjucentVertex)
          stack.append(adjucentVertex) 

customDict = {"a": ["b", "c"], #delcaration shows "c" is the deepest node to "a"
              "b": ["a", "d", "e"],
              "c": ["a", "e"],
              "d": ["b", "e", "f"],
              "e": ["b", "c", "d"],
              "f": ["d", "e"]}

simple_unweighted_undirected_graph = Graph(customDict)
print(simple_unweighted_undirected_graph.gdict)
simple_unweighted_undirected_graph.addEdge(vertex="e", edge="f")
print(simple_unweighted_undirected_graph.gdict["e"])

print("Traversing Through the graph with BFS")
simple_unweighted_undirected_graph.bfs("a")

print("Traversing Through the graph with DFS")
simple_unweighted_undirected_graph.dfs("a")
```

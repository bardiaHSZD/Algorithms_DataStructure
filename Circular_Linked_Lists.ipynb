```python
class Node:
  def __init__(self,value=None):
    self.value = value
    self.next = None

class circularLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def __iter__(self): #used for printing values and iterating
    node = self.head
    while node:
      yield node
      node = node.next
      if node == self.tail.next:
        break

  def createCircularLinkedList(self,nodeValue): #creation of circular single linked list     
    node = Node(nodeValue)
    node.next = node
    self.head = node
    self.tail = node
    return 'The circular single linked list has been created'

  def insertCircularLinkedList(self,value,location):
    if self.head is None:
      return 'The circular linked list does not exist'
    else:
      newNode = Node(value)

      if location == 0:
        newNode.next = self.head
        self.head = newNode
        self.tail.next = newNode

      elif location == -1:
        newNode.next = self.tail.next
        self.tail.next = newNode
        self.tail = newNode

      else:
        tempNode = self.head
        index = 0
        while (index<location-1):
          tempNode = tempNode.next
          index += 1
        nextNode=tempNode.next
        tempNode.next = nextNode
        newNode.next = nextNode
      return 'The node has been successfully inserted'  



circularLinkedListInstance = circularLinkedList()
circularLinkedListInstance.createCircularLinkedList(10.0)

print([node.value for node in circularLinkedListInstance]) 
circularLinkedListInstance.insertCircularLinkedList(15,0)
print([node.value for node in circularLinkedListInstance]) 
circularLinkedListInstance.insertCircularLinkedList(25,-1)
print([node.value for node in circularLinkedListInstance]) 
circularLinkedListInstance.insertCircularLinkedList(35,0)
print([node.value for node in circularLinkedListInstance]) 
```

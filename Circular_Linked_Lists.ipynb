```python
class Node:
  def __init__(self,value):
    self.value = value
    self.next = None

class SingleLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def __iter__(self):
    node = self.head
    while node:
      yield node
      node = node.next 

  def insert(self, value, location):
    newNode = Node(value)
    if self.head is None: # Caution: use 'is' instead of '=='
      self.head = newNode
      self.tail = newNode
    else:
      if location == 0:
        newNode.next = self.head
        self.head = newNode  
      elif location == -1:
        newNode.Next = None 
        self.tail.next = newNode
        self.tail = newNode
      else:
        tempNode = self.head
        index = 0
        while index < location - 1:
          tempNode = tempNode.next
          index += 1        
        nextNode = tempNode.next
        tempNode.next = newNode # current node is tempNode
        newNode.next = nextNode
        if tempNode == self.tail:
          self.tail = newNode
  def traverseLinkedList(self):
    if self.head is None:
      print('The single linked list does not exist!')
    else:
      tempNode = self.head
      while (tempNode is not None):
        print(tempNode.value)
        tempNode = tempNode.next
  def searchLinkedList(self,value):
    if self.head is None:
      print('The single linked list does not exist!') 
    else:
      tempNode = self.head
      while (tempNode is not None):
        if (tempNode.value == value):
          return 'Item found!'
        tempNode = tempNode.next 
    return 'Item not found!'
  def deleteNodeLinkedList(self,location):
    if self.head is None:
      print('The single linked list does not exist!') 
    else:
      if location == 0: #deleting the first node
        if self.head == self.tail: #only one node exists
          self.head = None
          self.tail = None
        else:
          self.head = self.head.next #we have more than one element
      elif location == -1: #deleting the last node
        if self.head == self.tail: #only one node exists
          self.head = None
          self.tail = None
        else:
          tempNode = self.head
          while (tempNode is not None):
            if tempNode.next == self.tail:
              break
            tempNode = tempNode.next
          tempNode.next = None
          self.tail = tempNode
      else:
          tempNode = self.head
          index = 0
          while (index < location - 1):
            tempNode = tempNode.next
            index += 1  
          nextNode = tempNode.next
          tempNode.next = nextNode.next   
  def deleteEntireLinkedList(self):
    if self.head is None:
      print('The single linked list does not exist!') 
    else:
      self.head = None
      self.tail = None



singleLinkedListInstance = SingleLinkedList()
node1 = Node(10.0)
node2 = Node(20.0)
singleLinkedListInstance.head = node1
singleLinkedListInstance.head.next = node2
singleLinkedListInstance.tail = node2

singleLinkedListInstance.insert(15.0,1)
singleLinkedListInstance.insert(25.0,1)
singleLinkedListInstance.insert(35.0,1)
singleLinkedListInstance.insert(45.0,1)

print([node.value for node in singleLinkedListInstance]) 

singleLinkedListInstance.deleteNodeLinkedList(2)
print([node.value for node in singleLinkedListInstance]) 

singleLinkedListInstance.deleteNodeLinkedList(0)
print([node.value for node in singleLinkedListInstance]) 

singleLinkedListInstance.deleteNodeLinkedList(-1)
print([node.value for node in singleLinkedListInstance]) 
singleLinkedListInstance.deleteEntireLinkedList()
print([node.value for node in singleLinkedListInstance])
```

### **Creation**

```python
class Node:
  def __init__(self,value=None):
    self.value = value
    self.next = None

class circularLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def __iter__(self): #used for printing values and iterating
    node = self.head
    while node:
      yield node
      if node.next == self.head:
        break
      node = node.next

  def createCircularLinkedList(self,nodeValue): #creation of circular single linked list     
    node = Node(nodeValue)
    node.next = node
    self.head = node
    self.tail = node
    return 'The circular single linked list has been created'


circularLinkedListInstance = circularLinkedList()
circularLinkedListInstance.createCircularLinkedList(10.0)

print([node.value for node in circularLinkedListInstance]) 
```

```python
class Node:
  def __init__(self, value=None):
    self.value = value
    self.next = None
    self.prev = None

class DoubleLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def __iter__(self):
    node = self.head
    while node:
      yield node
      node = node.next
  # Creation of double linked list
  def createDoubleLinkedList (self, nodeValue):
    node = Node(nodeValue)

    node.next = None
    node.prev = None  

    self.head = node
    self.tail = node 

    return 'The double linked list created successfully'

  def insertDoubleLinkedList(self,value,location):
    if self.head is None:
      return 'The double linked list does not exist'
    else:

      nextNode = Node(value)
      if location == 0:       

        nextNode.prev = None
        nextNode.next = self.head
        self.head.prev = nextNode
        self.head = nextNode

      elif location == -1:

        nextNode.next = None
        nextNode.prev = self.tail
        self.tail.next = nextNode
        self.tail = nextNode      
        
      else:

        index = 0
        tempNode = self.head
        while(index < location - 1):
          tempNode = tempNode.next
          index += 1
        nextNode.next = tempNode.next
        nextNode.prev = tempNode
        tempNode.next.prev = nextNode
        tempNode.next = nextNode

    return 'The inserstion was successfull'        



double_linked_list_instant = DoubleLinkedList()
double_linked_list_instant.createDoubleLinkedList(5.0)    

print([node.value for node in double_linked_list_instant])

double_linked_list_instant.insertDoubleLinkedList(10.0,0)
print([node.value for node in double_linked_list_instant])
double_linked_list_instant.insertDoubleLinkedList(11.0,0)
print([node.value for node in double_linked_list_instant])
double_linked_list_instant.insertDoubleLinkedList(12.0,0)
print([node.value for node in double_linked_list_instant])
double_linked_list_instant.insertDoubleLinkedList(15.0,-1)
print([node.value for node in double_linked_list_instant])
double_linked_list_instant.insertDoubleLinkedList(20.0,1)
print([node.value for node in double_linked_list_instant])
double_linked_list_instant.insertDoubleLinkedList(25.0,0)
print([node.value for node in double_linked_list_instant])
```

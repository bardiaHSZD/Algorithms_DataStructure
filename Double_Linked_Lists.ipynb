### **Creation**

```python
class Node:
  def __init__(self, value=None):
    self.value = value
    self.next = None
    self.prev = None

class DoubleLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def __iter__(self):
    node = self.head
    while node:
      yield node
      node = node.next
  # Creation of double linked list
  def createDoubleLinkedList (self, nodeValue):
    node = Node(nodeValue)

    node.next = None
    node.prev = None  

    self.head = node
    self.tail = node 

    return 'The double linked list created successfully' 

double_linked_list_instant = DoubleLinkedList()
double_linked_list_instant.createDoubleLinkedList(5)    

print([node.value for node in double_linked_list_instant])
```

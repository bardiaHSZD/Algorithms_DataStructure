## **Examples**

```python
newTuple = ('a', 'b', 'c', 'd', 'e')
print(newTuple[1])
print(newTuple[1:3])

# newTuple[1] = 'h' # Tuples are immutable so you cannot reassign values
#del newTuple[0] # Tuples are immutable so you cannot reassign values
```

```python
newTuple = ('a', 'b', 'c', 'd', 'e')

def searchTuple(pTuple, element):
  for i in pTuple:
    if i==element:
      return pTuple.index(i)
  return 'Not Found'    

searchTuple(newTuple,'e')

# Time complexity: O(n)
# Space complexity: O(1)
```

```python
myTuple = (1,4,3,2,5)
myTuple1 = (1,3,6,9,8,7,3)

print(myTuple + myTuple1) 
print(myTuple * 4)      
print(2 in myTuple1)

print(myTuple1.count(3))
print(myTuple1.index(9))
```

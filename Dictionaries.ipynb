# **Interview Questions**

**Q1. Missing Number:**

Description: Find the missing number between an array of integers, say 1 to 100.

```python
import numpy as np

# Hint: Summation of integers up to n is: n(n+1)/2
myList = [1, 2, 3, 4, 5, 6, 7, 8, 10]
def findMissing(missingArray):
  maximumValue = np.max(missingArray)
  expectedSum = maximumValue*(maximumValue+1)/2
  currentSum = np.sum(missingArray)
  missingValue = int(expectedSum - currentSum)
  return 'The missing value is: ' + str(missingValue)

findMissing(myList)  
```

## **Examples**

### **Updating/Assignment/Traversing/Linear Search:**

```python
myDict = {'name': 'Edy', 'age': 20}
myDict['age'] = 27 
# Time Complexity O(1).
# An extra memory is not required, so the space complexity is O(1)
print(myDict)

myDict['city'] = 'London'
# Time Complexity O(1).
# Space complexity is amortized O(1)
# If the complexity of dictionary is reached the space will be doubling
print(myDict)

def traverseDict(dict):
  for key in dict:
    print(key, dict[key])
traverseDict(myDict)   
# Time Complexity O(n).
# Space complexity is amortized O(1)

def searchDic(dict,value):
  for key in dict:
    if dict[key] == value:
      return key, value
  return 'The value does not exist!'    
searchDic(myDict, 27)
# Time Complexity O(n).
# Space complexity is amortized O(1)
```

### **Delete/Remove**

```python
myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}
myDict.pop('name')
print(myDict)

myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}
print(myDict.popitem()) # randomly selected
print(myDict)

myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}
myDict.clear() 
print(myDict)

myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}
del myDict['age']
print(myDict)

myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}
del myDict # even the dictionary itself
#print(myDict) # myDict is not defined

# Time Complexity: O(1); In the amortized case: O(n)
# Space Complexity: O(1)
```

### **Methods:**

```python
myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}

myDict2 = myDict.copy() #does not modify the original dictionary

print(myDict)
print(myDict2)

newDict = {}.fromkeys([1,2,3], 0)
print(newDict)
newDict = {}.fromkeys([1,2,3])
print(newDict)

print(myDict.get('age', 27))
print(myDict.get('age'))
print(myDict.get('address'))

print(myDict.items())

print(myDict.keys())

print(myDict.values())


print(myDict.popitem())
print(myDict)

myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}
print(myDict.setdefault('name', 'added'))
print(myDict.setdefault('name1', 'added')) # 'name1' not found 
print(myDict.items())

print(myDict.pop('name1', 'not'))
print(myDict.items())

myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}
newDict = {'a': 0, 'b': 1, 'age': 2}
myDict.update(newDict)
print(myDict)
```

### **Built-in Functions**

```python
myDict = {'name': 'Edy', 'age': 26, 'city': 'London', 'education': 'master'}

print('name' in myDict) # keys
print('Edy' in myDict.values()) # key 
# Unlike the list 'in' operator, dictionaries use hash tables so the time 
# complexity does not grow and it is O(1)

for key in myDict:
  print(key, myDict[key])
# Time complexity: O(n)
```

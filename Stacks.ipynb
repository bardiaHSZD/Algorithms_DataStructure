### **Operations using Lists(Pop/Push/Peek/isEmpty/Delete)**

```python
class Stack:
  def __init__(self):
    self.list = []
  
  def __str__(self):
    values = self.list.reverse()
    values = [str(x) for x in self.list]
    return '\n'.join(values)

#isEmpty: Time Complexity = O(1), Space Complexity = O(1)
  def isEmpty(self):
    if self.list == []:
      return True
    else:
      return False
    
#push: Time Complexity = Ammortized O(1), Space Complexity = O(1)
  def push(self,value):
    self.list.append(value)
    return "The element has been successfully inserted."

#push: Time Complexity = O(1), Space Complexity = O(1)
  def pop(self):
    if self.isEmpty():
      return "The stack is empty."    
    else:
      return self.list.pop()

#peek: Time Complexity = O(1), Space Complexity = O(1)
  def peek(self):
    if self.isEmpty():
      return "The stack is empty."    
    else:
      return self.list[len(self.list)-1]

#delete: Time Complexity = O(1), Space Complexity = O(1)
  def delete(self):
    self.list = None
    return "The stack was successfully deleted"

#isEmpty:
stackInstance = Stack()
print(stackInstance.isEmpty())          

#push:
stackInstance.push(1)
stackInstance.push(10)
stackInstance.push(100)
#print(stackInstance)

#pop:
stackInstance.pop()
#print(stackInstance)

#peek:
print(stackInstance.peek())

#delete:
print(stackInstance.delete())
```

### **Creation and operations with size limit (Pop/Push/Peek/isEmpty/Delete)**

```python
class Stack:
  def __init__(self, maxSize):
    self.maxSize = maxSize
    self.list = []
  def __str__(self):
    values = self.list.reverse()
    values = [str(x) for x in self.list]
    return "\n".join(values) 

#isEmpty: Time Complexity = O(1), Space Complexity = O(1)
  def isEmpty(self):
    if self.list == []:
      return True
    else:
      return False

#isFull: Time Complexity = O(1), Space Complexity = O(1) 
  def isFull(self):
    if len(self.list) == self.maxSize:
      return True
    else:
      return False  
    
#push: Time Complexity = Ammortized O(1), Space Complexity = O(1)
  def push(self,value):
    if self.isFull():
      return "The stack is full!"
    else:  
      self.list.append(value)
      return "The element has been successfully inserted."

#push: Time Complexity = O(1), Space Complexity = O(1)
  def pop(self):
    if self.isEmpty():
      return "The stack is empty."    
    else:
      return self.list.pop()

#peek: Time Complexity = O(1), Space Complexity = O(1)
  def peek(self):
    if self.isEmpty():
      return "The stack is empty."    
    else:
      return self.list[len(self.list)-1]

#delete: Time Complexity = O(1), Space Complexity = O(1)
  def delete(self):
    self.list = None
    return "The stack was successfully deleted"

#isEmpty:
stackInstance = Stack(4)
print(stackInstance.isEmpty())          

#push:
print(stackInstance.isFull())

#push:
stackInstance.push(1)
stackInstance.push(10)
stackInstance.push(100)
#print(stackInstance)

#pop:
stackInstance.pop()
#print(stackInstance)

#peek:
print(stackInstance.peek())

#delete:
print(stackInstance.delete())    
```

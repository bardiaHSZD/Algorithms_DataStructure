```python
# Fibonacci Series with Memoization
from time import process_time
def fibo(n):
  if n == 1:
    return 0
  if n == 2:
    return 1
  return fibo(n-1) + fibo(n-2)

def fiboMemo(n, memo):
  if n == 1:
    return 0
  if n == 2:
    return 1
  if n not in memo:
    memo[n] = fiboMemo(n-1, memo) + fiboMemo(n-2, memo)
  return memo[n]

t1_start = process_time() 
print(fibo(15))
t1_end = process_time() 
print('Fibonacci with memoization takes {:.6f} seconds'.format(t1_end-t1_start))

myDict = {}
t1_start = process_time() 
print(fiboMemo(15,myDict))
t1_end = process_time() 
print('Fibonacci with memoization takes {:.6f} seconds'.format(t1_end-t1_start))
```

```python
# Fibonacci Series with Memoization
from time import process_time
def fibo(n):
  if n == 1:
    return 0
  if n == 2:
    return 1
  return fibo(n-1) + fibo(n-2)

def fiboTab(n):
  tb = [0, 1]
  for i in range(2,n+1):  
    tb.append(fiboTab(i-1)+fiboTab(i-1))
  return tb(n-1)

def fiboMemo(n, memo):
  if n == 1:
    return 0
  if n == 2:
    return 1
  if n not in memo:
    memo[n] = fiboMemo(n-1, memo) + fiboMemo(n-2, memo)
  return memo[n]  

t1_start = process_time() 
print(fibo(15))
t1_end = process_time() 
print('Fibonacci takes {:.6f} seconds'.format(t1_end-t1_start))


t1_start = process_time() 
print(fibo(15))
t1_end = process_time() 
print('Fibonacci with tabulation takes {:.6f} seconds'.format(t1_end-t1_start))

myDict = {}
t1_start = process_time() 
print(fiboMemo(15,myDict))
t1_end = process_time() 
print('Fibonacci with memoization takes {:.6f} seconds'.format(t1_end-t1_start))
```

### **7. Number Factor Problem using Dynamic Programming**

image.png

```python
#Number factor problem using divide and conquer

def numberFactoDivCon(N):
  if N in (0,1,2):
    return 1
  elif N == 3:
    return 2
  else:
    return numberFactoDivCon(N-1) + numberFactoDivCon(N-3) + numberFactoDivCon(N-4) 

print(numberFactoDivCon(5))     
```

```python
#Number factor problem using dynamic programming

def numberFactoDynPrgTopDown(N, dp):
  if N in (0,1,2):
    return 1
  elif N == 3:
    return 2
  elif N not in dp:      
    dp[N] = numberFactoDynPrgTopDown(N-1,dp) + numberFactoDynPrgTopDown(N-3,dp) + numberFactoDynPrgTopDown(N-4,dp)
    return dp[N]
  else:   
    return dp[N]  

def numberFactoDynPrgDownTop(N):
  tb = [1, 1, 1, 2]
  for i in range(4, N+1):
    tb.append(numberFactoDynPrgDownTop(i-1)+numberFactoDynPrgDownTop(i-3)+numberFactoDynPrgDownTop(i-4))
  return tb[N]

myDict = {}
print(numberFactoDynPrgTopDown(5,myDict)) 
print(numberFactoDynPrgDownTop(5)) 
```
